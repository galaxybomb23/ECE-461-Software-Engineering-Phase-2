```python
import re
from collections import defaultdict
import json

logfile = '../logs/server_logs.log'

# Initialize dictionary
log_data = defaultdict(lambda: defaultdict(list))

# Regex patterns for endpoint and request
endpoint_pattern = r"--> /(.+)"
request_pattern = r"Request:"

def add_quotes_to_keys(json_string):
    # Regular expression to find keys without quotes
    pattern = r'(?<!")\b(\w+)\b(?=\s*:)'  # Matches keys before the colon
    # Replace keys with quoted keys
    corrected_json = re.sub(pattern, r'"\1"', json_string)
    return corrected_json
```

```python



# Variables to store current state
current_endpoint = None
current_method = None

# Open and read the log file
with open(logfile, "r") as log_file:
    while log_file.readline():
        #get new line
        line = log_file.readline()

        endpoint_match = re.search(endpoint_pattern, line)
        # Check if the line contains an endpoint
        if endpoint_match:
            
            # Store the endpoint
            current_endpoint, method = endpoint_match.group(1).split(": ", 1)
            print(current_endpoint, method)
            line = log_file.readline()
            request_match = re.search(request_pattern, line)
            
            if request_match:
                request = ""
                line = log_file.readline()
                while not line.startswith("12/") and not ("Ctx" in line):
                    request += line
                    line = log_file.readline()
                request = add_quotes_to_keys(request)
                print(request)
                request = json.loads(request)
                log_data[current_endpoint][method].append(request)
            
                

        





# Convert to a regular dictionary if needed
log_data = {key: dict(value) for key, value in log_data.items()}
# Print the result
print(log_data)
```
