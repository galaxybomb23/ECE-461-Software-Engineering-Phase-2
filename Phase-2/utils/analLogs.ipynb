```python
import re
from collections import defaultdict
import json

logfile = '../logs/server-all.log'

# Initialize dictionary
log_data = defaultdict(lambda: defaultdict(list))

# Regex patterns for endpoint and request
endpoint_pattern = r"--> /(.+): (\w+)"
request_pattern = r"Request\s*{"

# Open and read the log file
with open(logfile, 'r') as log_file:
    lines = log_file.readlines()

# Variables to store current state
current_endpoint = None
current_method = None
request = ''

# Iterate through each line in the log file
for line in lines:
    # Check for endpoint pattern
    endpoint_match = re.search(endpoint_pattern, line)
    if endpoint_match:
        current_endpoint = endpoint_match.group(1)
        current_method = endpoint_match.group(2)
        continue

    # Check for StartRequest pattern
    if 'StartRequest==>' in line:
        request = ''
        continue

    # Check for EndRequest pattern
    if '<==EndRequest' in line:
        if current_endpoint and current_method:

            log_data[current_endpoint][current_method].append(request.strip())
        current_endpoint = None
        current_method = None
        continue

    # Accumulate request data
    if current_endpoint and current_method:
        request += line.strip()

# Convert to a regular dictionary if needed
log_data = {key: dict(value) for key, value in log_data.items()}

# Print the result
print(json.dumps(log_data, indent=4))
```
